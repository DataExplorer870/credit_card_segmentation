Principal Component Analysis (PCA) is often used in clustering for dimensionality reduction, even when the reduced dimensions are still higher than two or three. Here are some reasons why PCA can be beneficial in clustering:

Noise Reduction: PCA helps in reducing noise in the data. It identifies and retains the most important components while discarding less informative ones. This can improve the quality of the clustering results by focusing on the essential patterns in the data.

Computational Efficiency: With a high number of original features, the computational complexity of clustering algorithms can increase significantly. PCA reduces the number of features, making clustering algorithms faster and more efficient..

Interpretability: Fewer dimensions make it easier to interpret the results. It's more manageable to understand and communicate the differences and relationships between clusters when dealing with a reduced set of components.

Enhanced Cluster Separation: By focusing on the most informative components, PCA can enhance the separation between clusters. This is particularly useful when clusters are not well-separated in the original feature space.

```python
pca = PCA()
X_pca = pca.fit_transform(df_scaled)
```

```python
df_pca = pd.DataFrame(data = X_pca, columns = [f'PC{i+1}' for i in range(X_pca.shape[1])])
```

```python
df_pca.head()
```

```python
evr = list(pca.explained_variance_ratio_)
```

```python
evr = sorted(evr, reverse=True)
```

```python
# Cummulative explained variance ratio
evr_cum = list(pca.explained_variance_ratio_.cumsum())
```

```python
num_components = [x + 1 for x in range(X_pca.shape[1])]
```

```python
fig, ax = plt.subplots(1, 2, figsize = (12, 5))

sns.barplot(x = num_components, y = evr, ax = ax[0])
ax[0].set_xlabel("Component")
ax[0].set_title("% Explained Variance")

sns.lineplot(x = num_components, y = evr_cum, marker = 'o' ,ax = ax[1])
ax[1].set_xlabel("Component")
ax[1].set_title("% Cumulative Variance")

fig.show()
```
